```python
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv(r"C:\Users\asus\Downloads\archive (4)\movies.csv")
```

```python
df.head()
```

```python
# to sea if there is any missing data
for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print('{}  - {}%'.format(col,pct_missing))
```

```python
df=df.dropna()
```

```python
df
```

```python
# Data types for our columns
df.dtypes
```

```python
#change type of columns
df['budget']= df['budget'].round().astype('int64')

df['gross']= df['gross'].round().astype('int64')
```

```python
df
```

```python
df=df.sort_values(by=['gross'],inplace=False, ascending = False)
```

```python
#drop any duplicates
df['company'].drop_duplicates().sort_values(ascending= False)
```

```python
df
```

```python
# Scattor plot budget vs gross

%matplotlib inline

plt.scatter(x = df['budget'],y=df['gross'])
plt.title(["Budget vs Gross Earnings"])
plt.xlabel("Budget")
plt.ylabel("Gross")
plt.show()
```

```python
correlation_matrix = df.corr(numeric_only = True , method ='pearson')
sns.heatmap(correlation_matrix , annot= True)
plt.title(["correlation matrix for numeric features"])
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
df = df_numarized

for col_name in df_numarized.columns:
    if df_numarized[col_name].dtype == 'object':  
        df_numarized[col_name] = df_numarized[col_name].astype('category')
        df_numarized[col_name] = df_numarized[col_name].cat.codes
        
df_numarized
```
